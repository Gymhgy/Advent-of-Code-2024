```python
inp = open('day14.txt', 'r').read()
```

```python
import re
from functools import reduce
arr = [0] * 4
for line in inp.split('\n'):
    x,y,vx,vy = [int(d) for d in re.findall(r"-?\d+", line)]
    x += vx * 100
    y += vy * 100
    x %= 101
    y %= 103
    if x == 101 // 2 or y == 103 // 2:
        continue
    arr[(x // 51) * 2 + (y // 52)] += 1

reduce(lambda a,b: a*b, arr, 1), len(inp.split('\n'))
```

```python
bots = [tuple(int(d) for d in re.findall(r"-?\d+", line)) for line in inp.split('\n')]

for i in range(0, 10403):
    state = set(((x + vx * i)%101, (y + vy * i)%103) for x,y,vx,vy in bots)
    
    max_count = 0

    for (x1, y1) in state:
        cnt = 1
        for x2 in range(x1 + 1, 101):
            if (x2, y1) in state:
                cnt += 1
            else:
                break
        max_count = max(max_count, cnt)

    if max_count > 10:
        print(i)
        for x in range(101):
            for y in range(103):
                if (x,y) in state:
                    print('#', end='')
                else:
                    print('.', end='')
            print()
        print()
        print()
```
