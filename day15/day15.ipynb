```python
inp = open('day15.txt', 'r').read()
```

```python
grid, inst = inp.split('\n\n')
grid = grid.split()
x = [i for i in range(len(grid)) if '@' in grid[i]][0]
y = grid[x].index('@')

grid = [list(row) for row in grid]
inst = inst.replace('\n', '')
grid[x][y] = '.'
dir = {"<": (0, -1), ">": (0, 1), "^": (-1, 0), "v": (1, 0)}
for move in inst:
    dx, dy = dir[move]
    nx, ny = x + dx, y + dy
    if grid[nx][ny] == '.':
        x, y = nx, ny
    elif grid[nx][ny] == 'O':
        sx, sy = nx, ny
        while grid[nx][ny] == 'O':
            nx += dx
            ny += dy
        if grid[nx][ny] == '.':
            grid[sx][sy] = '.'
            grid[nx][ny] = 'O'
            x, y = sx, sy

cnt = 0
for x, l in enumerate(grid):
    for y, c in enumerate(l):
        if c == 'O':
            cnt += 100 * x + y

cnt
```

```python
grid, inst = inp.split('\n\n')
grid = grid.replace('#', '##').replace('O', '[]') \
           .replace('.', '..').replace('@', '@.')
grid = grid.split()
x = [i for i in range(len(grid)) if '@' in grid[i]][0]
y = grid[x].index('@')

from copy import deepcopy
grid = [list(row) for row in grid]
def print_grid(grid):
    global x, y
    for i, row in enumerate(grid):
        if i == x:
            t = list(row)
            t[y] = '@'
            print(''.join(t))
        else:
            print("".join(row))
    print()

inst = inst.replace('\n', '')
grid[x][y] = '.'
dir = {"<": (0, -1), ">": (0, 1), "^": (-1, 0), "v": (1, 0)}
for move in inst:
    dx, dy = dir[move]
    nx, ny = x + dx, y + dy
    if grid[nx][ny] == '.':
        x, y = nx, ny
    elif grid[nx][ny] in '[]':
        boxes = set()
        st = [(nx, ny)]
        can_push = True
        new_grid = deepcopy(grid)
        while st:
            sx, sy = st.pop()
            if (sx, sy) in boxes:
                continue
            boxes.add((sx, sy))
            other = (sx, sy+1) if grid[sx][sy] == '[' else (sx, sy-1)
            st.append(other)

            ssx = sx + dx
            ssy = sy + dy
            
            if grid[ssx][ssy] in '[]':
                st.append((ssx, ssy))
            elif grid[ssx][ssy] == '#':
                can_push = False
                break

        if not can_push: continue
        for bx, by in boxes:
            new_grid[bx][by] = '.'
        for bx, by in boxes:
            new_grid[bx + dx][by + dy] = grid[bx][by]
        grid = new_grid
        x, y = nx, ny


cnt = 0
for x, l in enumerate(grid):
    for y, c in enumerate(l):
        if c == '[':
            cnt += 100 * x + y

cnt
```
