```python
inp = open('day24.txt', 'r').read()
```

```python
init, gates = inp.split('\n\n')
init = {n: int(v) for n, v in map(lambda l: l.split(': '), init.split('\n'))}

gates = gates.split('\n')
og = gates
zs = set(l.split()[-1] for l in gates if l.split()[-1][0] == 'z')
while any(z not in init for z in zs):
    next = []
    for g in gates:
        a, op, b, _, r = g.split()
        if a not in init or b not in init:
            next.append(g)
            continue
        if op == 'XOR':
            init[r] = init[a] ^ init[b]
        if op == 'AND':
            init[r] = init[a] & init[b]
        if op == 'OR':
            init[r] = init[a] | init[b]
    gates = next
    

i = 0
res = ""
while True:
    if 'z' + str(i).rjust(2, '0') in init:
        res = str(init['z' + str(i).rjust(2, '0')]) + res
        i += 1
    else:
        break
    

int(res, 2)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import defaultdict

G = nx.DiGraph()
alt = {}
for g in og:
    a, op, b, _, r = g.split()
    if a[0] == 'x' or a[0] == 'y':
        if b[0] == 'x' or b[0] == 'y':
            alt[r] = f"({r}={a} {op} {b})"
for g in og:
    a, op, b, _, r = g.split()
    if r not in G:
        G.add_node(r)
    G.add_edge(a, r)
    G.add_edge(b, r)
    a = alt.get(a, a)
    b = alt.get(b, b)
    G.nodes[r]['m'] = f"{a} {op} {b}"


for node in list(G.nodes):
    if node in alt or node[0] == 'x' or node[0] == 'y':
        G.remove_node(node)

for layer, nodes in enumerate(nx.topological_generations(G)):
    for node in nodes:
        G.nodes[node]["layer"] = -layer

pos = nx.multipartite_layout(G, subset_key="layer", align="horizontal")
for a, b in pos.items():
    if a[0] == 'x' or a[0] == 'y':
        b[0] *= 2
    else:
        b[0] *= 10
plt.figure(figsize=(50, 150))

nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=2000, font_size=12, font_weight="bold", arrowsize=20)

node_labels = {node: f"\n\n{data.get('m', '')}" for node, data in G.nodes(data=True)}
nx.draw_networkx_labels(G, pos, labels=node_labels)

plt.show()
```

```python
dag = {}
for g in og:
    a, op, b, _, r = g.split()
    dag[r] = (a, op, b)
def s(p, depth=2):
    if p[0] == 'x' or p[0] == 'y':
        return
    print(p, dag[p])
    if depth > 0:
        s(dag[p][0], depth-1)
        s(dag[p][2], depth-1)

s('z02', 3)
```
